## **1. Working With Date**

```python
date.info()
```

```python
date['date']=pd.to_datetime(date['date'])
date.info()
```

```python
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
# Extracting Year
date['Date_Year']=date['date'].dt.year
date.head()
```

```python
# Extracting Month
date['Date_Month']=date['date'].dt.month
date.head()
```

```python
# Extracting Day
date['Date_Day']=date['date'].dt.day
date.head()
```

```python
# Extracting Weak_of_Year
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
# Extracting Name of Day
date['Date_Day_Name']=date['date'].dt.day_name()
date.head()
```

```python
# Extracting Name of Month
date['Date_Month_Name']=date['date'].dt.month_name()
date.head()
```

```python
# Extracting Day_of_Year
date['Day_of_Year']=date['date'].dt.day_of_year
date.head()
```

```python
# Extracting Quarter
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
# Extracting Semaster
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
date.head()
```

## **2. Working With Time**

```python
time.info()
```

```python
time['date']=pd.to_datetime(time['date'])
time.info()
```

```python
time.drop(columns=['msg'],inplace=True)
```

```python
# Extracting hour,min and seconds
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

```python
# Extract Time Part
time['Time']=time['date'].dt.time
time.head()
```

```python
# Extracting Time elapsed between dates
import datetime
today=datetime.datetime.today()

# in days
today-time['date']
```

```python
# in seconds
(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minuts
(today-time['date'])/np.timedelta64(1,'m')
```

```python
# in hours
(today-time['date'])/np.timedelta64(1,'h')
```
