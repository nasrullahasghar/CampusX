# **Integer Encoding**

```python
docs = [
    'Pakistan zindabad',
    'Pakistan Pakistan',
    'dil se dil tak',
    'jeetega bhai jeetega Pakistan jeetega',
    'quaid-e-azam zindabad',
    'Babar Babar',
    'wasim wasim',
    'shahid bhai',
    'imran khan ki zindabad',
    'Pakistan payindah baad'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token="<nothing>")
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_docs
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_index
```

```python
vocab_size = len(tokenizer.word_index) + 1  
print(vocab_size)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
padded_sequences = pad_sequences(sequences,padding = 'post')
padded_sequences
```

```python
from keras import Sequential
from keras.layers import Embedding,Dense
```

```python
model = Sequential()
model.add(Embedding(vocab_size,output_dim=2,input_length=5))

model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
pred = model.predict(padded_sequences)  # Use padded_sequences here
print(pred)
```

# **2. Sentiment Analysis**

```python
import tensorflow
from tensorflow import keras
from keras.datasets import imdb
from keras.layers import Dense,Embedding,SimpleRNN,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train.shape
```

```python
X_train[0:2]
```

```python
len(X_train[0])
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
word_index = imdb.get_word_index()
len(word_index)
```

```python
unique_words = len(word_index) + 3 

print("Number of unique words in the IMDB dataset:", unique_words)
```

```python
model = Sequential()
model.add(Embedding(unique_words,output_dim=2,input_length=50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
