### **1. Agglomerative Hierarchical Clustring**

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('./data/hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
df.head()
```

```python
df.shape
```

```python
data = df.iloc[:,3:5].values
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=5)
labels_ = model.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=[10,6])
plt.scatter(data[:,0],data[:,1],c=model.labels_,cmap='rainbow')
```
